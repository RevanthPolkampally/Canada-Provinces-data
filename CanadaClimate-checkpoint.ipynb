```python
#fill null vales with median as these are numerical values
def imp_median(series):
    return series.fillna(series.median())
```

```python
data.MEAN_TEMPERATURE_CALGARY= data["MEAN_TEMPERATURE_CALGARY"].transform(imp_median)
data.TOTAL_PRECIPITATION_CALGARY= data["TOTAL_PRECIPITATION_CALGARY"].transform(imp_median)
data.MEAN_TEMPERATURE_EDMONTON= data["MEAN_TEMPERATURE_EDMONTON"].transform(imp_median)
data.TOTAL_PRECIPITATION_EDMONTON= data["TOTAL_PRECIPITATION_EDMONTON"].transform(imp_median)
data.MEAN_TEMPERATURE_HALIFAX= data["MEAN_TEMPERATURE_HALIFAX"].transform(imp_median)
data.TOTAL_PRECIPITATION_HALIFAX= data["TOTAL_PRECIPITATION_HALIFAX"].transform(imp_median)
data.MEAN_TEMPERATURE_MONCTON= data["MEAN_TEMPERATURE_MONCTON"].transform(imp_median)
data.TOTAL_PRECIPITATION_MONCTON= data["TOTAL_PRECIPITATION_MONCTON"].transform(imp_median)
data.MEAN_TEMPERATURE_MONTREAL= data["MEAN_TEMPERATURE_MONTREAL"].transform(imp_median)
data.TOTAL_PRECIPITATION_MONTREAL= data["TOTAL_PRECIPITATION_MONTREAL"].transform(imp_median)
data.MEAN_TEMPERATURE_OTTAWA= data["MEAN_TEMPERATURE_OTTAWA"].transform(imp_median)
data.TOTAL_PRECIPITATION_OTTAWA= data["TOTAL_PRECIPITATION_OTTAWA"].transform(imp_median)
data.MEAN_TEMPERATURE_QUEBEC= data["MEAN_TEMPERATURE_QUEBEC"].transform(imp_median)
data.TOTAL_PRECIPITATION_QUEBEC= data["TOTAL_PRECIPITATION_QUEBEC"].transform(imp_median)
data.MEAN_TEMPERATURE_SASKATOON= data["MEAN_TEMPERATURE_SASKATOON"].transform(imp_median)
data.TOTAL_PRECIPITATION_SASKATOON= data["TOTAL_PRECIPITATION_SASKATOON"].transform(imp_median)
data.MEAN_TEMPERATURE_STJOHNS= data["MEAN_TEMPERATURE_STJOHNS"].transform(imp_median)
data.TOTAL_PRECIPITATION_STJOHNS= data["TOTAL_PRECIPITATION_STJOHNS"].transform(imp_median)
data.MEAN_TEMPERATURE_TORONTO= data["MEAN_TEMPERATURE_TORONTO"].transform(imp_median)
data.TOTAL_PRECIPITATION_TORONTO= data["TOTAL_PRECIPITATION_TORONTO"].transform(imp_median)
data.MEAN_TEMPERATURE_VANCOUVER= data["MEAN_TEMPERATURE_VANCOUVER"].transform(imp_median)
data.TOTAL_PRECIPITATION_VANCOUVER= data["TOTAL_PRECIPITATION_VANCOUVER"].transform(imp_median)
data.MEAN_TEMPERATURE_WHITEHORSE= data["MEAN_TEMPERATURE_WHITEHORSE"].transform(imp_median)
data.TOTAL_PRECIPITATION_WHITEHORSE= data["TOTAL_PRECIPITATION_WHITEHORSE"].transform(imp_median)
data.MEAN_TEMPERATURE_WINNIPEG= data["MEAN_TEMPERATURE_WINNIPEG"].transform(imp_median)
data.TOTAL_PRECIPITATION_WINNIPEG= data["TOTAL_PRECIPITATION_WINNIPEG"].transform(imp_median)
```

```python
def date_convert(date_to_convert):
     return datetime.strptime(date_to_convert, '%d %m, %Y').strftime('%d-%m-%Y')

data['LOCAL_DATE'] = data['LOCAL_DATE'].apply(date_convert)
```
